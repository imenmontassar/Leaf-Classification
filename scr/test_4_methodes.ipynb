# II) Classification par la méthode des forêts aléatoires : 

## a: méthode simple avec entrainement et prédiction sans cross validation ni prétraitement

```python
from forets_aleatoires import Forets_Aleatoires

# préparation des données

x_train=df_train.drop(["species","id"], axis=1)
x_etiquette=df_train["species"]

# application du modèle
model_foret_alea=Forets_Aleatoires(x_train,x_etiquette)
model_foret_alea.entrainement()
model_foret_alea.prediction()
model_foret_alea.resultats()
```

## b: méthode avec selection selon la variance (PCA) avec entrainement et prédiction sans cross validation

```python
# prétraitement des données avec PCA en conservant 90% de l'information selon le critère de la variance
pretrait=Pretraitement(df_train)
pretrait.selection_avec_PCA(0.90)
# application du modèle
model_foret_alea=Forets_Aleatoires(pretrait.x_train,x_etiquette,False)
model_foret_alea.entrainement()
model_foret_alea.prediction()
model_foret_alea.resultats()
```

```python
# prétraitement des données avec PCA en conservant 95% de l'information selon le critère de la variance
pretrait=Pretraitement(df_train)
pretrait.selection_avec_PCA(0.95)
# application du modèle
model_foret_alea=Forets_Aleatoires(pretrait.x_train,x_etiquette,False)
model_foret_alea.entrainement()
model_foret_alea.prediction()
model_foret_alea.resultats()
```

```python
# prétraitement des données avec PCA en conservant 99% de l'information selon le critère de la variance
pretrait=Pretraitement(df_train)
pretrait.selection_avec_PCA(0.99)
# application du modèle
model_foret_alea=Forets_Aleatoires(pretrait.x_train,x_etiquette,False)
model_foret_alea.entrainement()
model_foret_alea.prediction()
model_foret_alea.resultats()
```

## c: méthode simple avec entrainement et prédiction avec cross validation mais sans prétraitement

```python
#test du modèle avec la validation croisée 
model_foret_alea=Forets_Aleatoires(pretrait.x_train,x_etiquette,True)
model_foret_alea.entrainement()
model_foret_alea.prediction()
model_foret_alea.resultats()
```

## d: meilleurs paramêtres 

```python
# application du modèle
model_foret_alea=Forets_Aleatoires(x_train,x_etiquette)
model_foret_alea.nmb_arbre = 500
model_foret_alea.criterion = 'entropy' # gini et entropy
model_foret_alea.min_samples_split = 4
model_foret_alea.entrainement()
model_foret_alea.prediction()
model_foret_alea.resultats()
```
